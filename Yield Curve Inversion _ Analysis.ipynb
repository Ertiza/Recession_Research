# Unemployment rate% Data - Fed

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt

# Your FRED API Key
FRED_API_KEY = "c15723a69d8fb079ddd6876960cab524"

# Unemployment Rate (UNRATE) - U.S.
FRED_UNEMPLOYMENT_URL = f"https://api.stlouisfed.org/fred/series/observations?series_id=UNRATE&api_key={FRED_API_KEY}&file_type=json"

# Fetch the data
response = requests.get(FRED_UNEMPLOYMENT_URL)
data = response.json()

# Convert to DataFrame
unemployment_df = pd.DataFrame(data["observations"])
unemployment_df["date"] = pd.to_datetime(unemployment_df["date"])
unemployment_df["unemployment_rate"] = unemployment_df["value"].astype(float)

# Save to Excel
unemployment_file = "Unemployment_Rate.xlsx"
unemployment_df.to_excel(unemployment_file, sheet_name="Unemployment Data", index=False)

print(f"Unemployment data successfully saved to {unemployment_file}")

# Plot Unemployment Rate
plt.figure(figsize=(12, 5))
plt.plot(unemployment_df["date"], unemployment_df["unemployment_rate"], label="Unemployment Rate", color='red')
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.title("U.S. Unemployment Rate Over Time")
plt.legend()
plt.grid()
plt.show()
```
